```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
pymaceuticals_complete_data = pd.merge(study_results, mouse_metadata, how="left", on=["Mouse ID", "Mouse ID"])
pymaceuticals_complete_data

# Display the data table for preview
```

```python
# Checking the number of mice.
mouse_count= pymaceuticals_complete_data['Mouse ID'].nunique()
mouse_count
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
#duplicate_mouseID = pymaceuticals_complete_data['Mouse ID'].duplicated()
#duplicate_mouseID

duplicate_id=pymaceuticals_complete_data.loc[pymaceuticals_complete_data.duplicated(subset=['Mouse ID', 'Timepoint']),'Mouse ID'].unique()
duplicate_id

```

```python
dup_row_series = pymaceuticals_complete_data.duplicated(subset=['Mouse ID', 'Timepoint'], keep=False)

duplicate_mouseIDs = pymaceuticals_complete_data[dup_row_series]["Mouse ID"].unique()
duplicate_mouseIDs
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
dup_mouse_series = pymaceuticals_complete_data["Mouse ID"].isin(duplicate_mouseIDs)
pymaceuticals_complete_data[dup_mouse_series]
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
#clean_data = pymaceuticals_complete_data.drop_duplicates(subset =['Mouse ID'], keep= False)
clean_df = pymaceuticals_complete_data[~dup_mouse_series]
clean_df
```

```python
# Checking the number of mice in the clean DataFrame
mouse_count= clean_df['Mouse ID'].nunique()
mouse_count
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, 
#and SEM of the tumor volume for each regimen
#clean_df.groupby("Drug Regimen").mean()["Tumor Volume (mm3)"]

mean = clean_df['Tumor Volume (mm3)'].groupby(clean_df['Drug Regimen']).mean()
median = clean_df['Tumor Volume (mm3)'].groupby(clean_df['Drug Regimen']).median()
var = clean_df['Tumor Volume (mm3)'].groupby(clean_df['Drug Regimen']).var()
std = clean_df['Tumor Volume (mm3)'].groupby(clean_df['Drug Regimen']).std()
sem = clean_df['Tumor Volume (mm3)'].groupby(clean_df['Drug Regimen']).sem()

summary_stat = pd.DataFrame({"Mean Tumor Volume":mean, 
                            "Median Tumor Volume":median, 
                           "Tumor Volume Variance":var, 
                           "Tumor Volume Std. Dev.":std, 
                           "Tumor Volume Std. Err.":sem})
# Display the Summary statistics table grouped by 'Drug Regimen' column
summary_stat
    



```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)
aggfunctions = {"Tumor Volume (mm3)": ["mean", "median", "var", "std", "sem"]}

clean_df.groupby("Drug Regimen").agg(aggfunctions)

# Using the aggregation method, produce the same summary statistics in a single line
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  

# Capomulin_df = clean_df.loc[clean_df["Drug Regimen"] == "Capomulin",:]
# Ramicane_df = clean_df.loc[clean_df["Drug Regimen"] == "Ramicane", :]
# Infubinol_df = clean_df.loc[clean_df["Drug Regimen"] == "Infubinol", :]
# Ceftamin_df = clean_df.loc[clean_df["Drug Regimen"] == "Ceftamin", :]
# Start by getting the last (greatest) timepoint for each mouse

max_df = clean_df.groupby("Mouse ID").max()[["Timepoint"]].reset_index()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
merge_df = pd.merge(max_df, clean_df, how="left", on=["Mouse ID", "Timepoint"])
merge_df
```

```python
treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
# Calculate the final tumor volume of each mouse across four of the treatment regimens:

#Capomulin, Ramicane, Infubinol, and Ceftamin

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_data = []
#loop thru the promising drugs
for treatment in treatments:
   
    # Locate the rows which contain mice on this drug and get the tumor volumes as a series
    tumor_vol = merge_df.loc[merge_df["Drug Regimen"] == treatment]["Tumor Volume (mm3)"]
   # add the series of tumor volume on this drug to the ploting list   
    tumor_vol_data.append(tumor_vol)
    
   # Calculate the IQR for this drug   

    q1 = tumor_vol.quantile(.25)
    q3 = tumor_vol.quantile(.75)
    IQR = q3-q1
    upper_vol=q3+(1.5*IQR)
    lower_vol=q1-(1.5*IQR)
    
     # Determine outliers using upper and lower bounds for this drug
    outliers= tumor_vol.loc[(tumor_vol<lower_vol)|(tumor_vol>upper_vol)]
    print(treatment, "'s potential outliers:", outliers)
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
plt.boxplot(tumor_vol_data)
plt.xlabel("Drug Regimen")
plt.ylabel("Final Tumor Volume (mm3")



plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
Capomulin_df = clean_df.loc[clean_df["Drug Regimen"] == "Capomulin"]
Capomulin_agg_df = Capomulin_df.groupby("Mouse ID").mean()[["Tumor Volume (mm3)", "Weight (g)"]]
plt.scatter(Capomulin_agg_df["Weight (g)"],Capomulin_agg_df["Tumor Volume (mm3)"])
```

```python
plt.scatter(Capomulin_agg_df["Weight (g)"],Capomulin_agg_df["Tumor Volume (mm3)"])
slope, intercept, r_value, p_value, std_err = st.linregress(Capomulin_agg_df["Weight (g)"],Capomulin_agg_df["Tumor Volume (mm3)"])
print("r-squared:", r_value)
plt.plot(Capomulin_agg_df["Weight (g)"], intercept + slope*Capomulin_agg_df["Weight (g)"], 'r', label='fitted line')
plt.xlabel("Weight (g)")
plt.ylabel("Tumor Volume (mm3)")
plt.show()    
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
plt.scatter(Capomulin_agg_df["Weight (g)"],Capomulin_agg_df["Tumor Volume (mm3)"])
slope, intercept, r_value, p_value, std_err = st.linregress(Capomulin_agg_df["Weight (g)"],Capomulin_agg_df["Tumor Volume (mm3)"])
print("r-squared:", r_value)
plt.plot(Capomulin_agg_df["Weight (g)"], intercept + slope*Capomulin_agg_df["Weight (g)"], 'r', label='fitted line')
plt.xlabel("Weight (g)")
plt.ylabel("Tumor Volume (mm3)")
plt.show()    
```
